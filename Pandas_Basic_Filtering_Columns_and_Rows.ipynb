### FILTER BASED OFF OF THE INDEX (FILTER)

- set a specific column as index
  - df.set_index('Column_name')

- filter specific columns you want to keep
  - df.filter(items=['column1','column2'])
  - df.filter(like='column')
  - df2.filter(regex='column1|column2', axis = 0) * 'like' parameter in 'filter()' method of pandas does not accept a list; instead use regex pattern
  - axis=0 / axis=1 (speficy which axis you want to search on / axis=o â¬‡, axis=1 âž¡)



#### Regular Expression Basics
1) | : Acts as an OR operator.
  - Example: United|Zim matches any string containing either "United" or "Zim".  

2) ^ : Matches the beginning of a string.
  - Example: ^United matches strings starting with "United".  

3) ðŸ’² : Matches the end of a string.
  - Example: Zim$ matches strings ending with "Zim".  

4) . : Matches any single character.  
  - Example: U.it matches "Unit" but not "Unite".

5) [ ] : Matches any one character inside the brackets.
  - Example: [UZ]im matches "Zim" or "Uim".

```python
df2=df.set_index('Country')
df2
```

```python
df2.filter(items=['Continent', 'CCA3'])
```

```python
df2.filter(items=['Continet', 'CCA3'], axis=0)
```

```python
df2.filter(items=['Continent', 'CCA3'],axis=1)
```

```python
df2.filter(items=['Zimbabwe'], axis=0)
```

```python
df2.filter(like='United', axis=0)
```

```python
df2.filter(regex='United|Zim', axis = 0)
```

### ORDER BY

```python
df[df['Rank']<10].sort_values(by='Rank')
```

```python
df[df['Rank']<10].sort_values(by='Rank', ascending=False)
```

```python
df[df['Rank']<10].sort_values(by=['Continent', 'Rank'],ascending=True)
```

```python
df[df['Rank']<10].sort_values(by=['Continent', 'Rank'],ascending=[False,True])
```
