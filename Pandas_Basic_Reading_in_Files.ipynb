## 1. IMPORT CSV FIELS

```python
from google.colab import files
uploaded = files.upload()
```

### WHY to import **io**?

- files.upload() : saves the file in **binary** format  
                    - **binary** means the content is stored as **a mix of numbers and letters** that is not human-readable
- io.BytesIO : a tool that helps **convert** this "strange" format into something that can be read

```python
import pandas as pd
import io
```

```python
df = pd.read_csv(io.BytesIO(uploaded['countries of the world.csv']))
df.head()
```

### INDEX

When reading files in Python, it automatically generates index column.  

### HEADER  

- When reading fiels in Python, it takes the first row as header.  

- You can change headers like [header=None]  -> hearder will have index  

- You can even specify names of headers like [header=None, names=['Country','Region']

```python
df1 = pd.read_csv(io.BytesIO(uploaded['countries of the world.csv']), header=None)
df1.head()
```

```python
df2 = pd.read_csv(io.BytesIO(uploaded['countries of the world.csv']), header=None, names=['Country','Region'])
df2.head()
```

####6-4. looking at a specific cell using **df.loc**
  - df.loc : using labels (names), choose row and column
    - df.loc[row#]: access all data for the row
    - df.loc[row#, 'column_name']: access a specific value from a particular row and column
    - df.loc[row1:row2, 'column_name']: access a range of rows for a specific column
    - df.loc[row1:row2, ['column1', 'column2']]: access multiple columns for a range of rows

####6-5. looking at a specific cell using **df.iloc**

  - df.iloc : using index positions (integers), you can choose rows and columns
    - df.iloc[row#]: access all data for the row based on the index position
    - df.iloc[row#, column#]: access a specific value using index positions (both row and column)
    - df.iloc[row1:row2, column#]: access a range of rows for a specific column (based on index positions)
    - df.iloc[row1:row2, [column1, column2]]: access multiple columns for a range of rows using index positions

```python
print(df.iloc[1])  # All data for row index 1 (second row)
print(df.iloc[1, 2])  # Specific data for row index 1, column index 2 ('Albania')
print(df.iloc[0:3, 0])  # Multiple data from rows 0, 1, 2, 'Rank' column (index 0)
print(df.iloc[0:3, [0, 2]])  # Multiple data from rows 0, 1, 2, 'Rank' and 'Country' columns (index 0 and 2)
```
